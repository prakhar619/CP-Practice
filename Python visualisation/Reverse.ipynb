```python
def reverse(x):
    ans = 0
    while(x > 0):
        t = x % 10
        ans = ans * 10 + t
        x = int(x / 10)
    return ans
```

```python
llist = [i for i in range(1000,1020)]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev = list(map(reverse,llist))
```

```python
plt.plot(llist,rev)
```
